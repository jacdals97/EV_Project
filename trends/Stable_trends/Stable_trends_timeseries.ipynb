# Analysis of candidate reference terms
Time series analysis to establish reference term to be compared in relation to the trends of electric vehicles. 

By queriying the unofficial Google Trends API to return the geographic (city-wise) trends of more than two terms you get the proportion of search volume of each term out of the total search volume of the terms in the city during your specified time period.

By selecting a clever reference term, this enables us to get a proper time series of the electric vehicle interest in danish cities albeit with some assumptions. We are looking for a term which trend is stationary ie the mean of the trend is constant and so is its variance. If this is the case then on average, the change in the proportion of electric vehicle searches should only be affected by the volume of electric vehicle searches.

We initially brainstormed ideas for stationary terms and came up with terms such as 'weather' however we quickly saw that our candidate reference terms can't be too popular either because they will dwarf the electric vehicle trend by shrinking its proportion to zero as the API only returns integer values for the proportion.

We subsequently followed some of Google own search tips at https://blog.google/products/search/15-tips-getting-most-out-google-trends/ and came up with the following candidate list by looking at the Regional Trends of the terms.

Candidates include: Sports, Movies, News, Weather and Music

The time series that will be analysed in this section only include data from regional data i.e the regions of Denmark (https://en.wikipedia.org/wiki/Regions_of_Denmark), we thus assume that if we can establish stationarity of these time series then that stationarity also holds for the time series of individual cities.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import joypy

from pytrends.request import TrendReq

# stationarity tests
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss

# autocorrelation plots
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# create a datetime object from a string
low = datetime.strptime('2017-01-01 00:00:00', '%Y-%m-%d %H:%M:%S')
high = datetime.strptime('2018-01-01 00:00:00', '%Y-%m-%d %H:%M:%S')

low = pd.Timestamp(low)
high = pd.Timestamp(high)
```

### Build dictionaries and loop through to generate subsettable dataset

```python
DK_geos = {'Nordjylland':"DK-81", 'Hovedstaden':"DK-84", 'Midtjylland':"DK-82", 'Sjælland':"DK-85", 'Syddanmark':"DK-83"}
topics = {'News': '/m/05jhg', 'Movies':'/m/02vxn', 'Sports':'/m/06ntj', 'Music':'/m/04rlf', 'weather':'/m/0866r'}

pytrends = TrendReq(hl='en-US', tz=-60, timeout=(15,30), retries=3, backoff_factor=0.2)

topic_trends = pd.DataFrame()
for geo, geo_code in DK_geos.items():
    for topic, topic_code in topics.items():
        pytrends.build_payload(kw_list = [topic_code], geo= geo_code, timeframe= 'today 5-y', gprop='')
        df = pytrends.interest_over_time()
        df['Region'] = [geo]*len(df)
        df = df.rename(columns={topic_code: 'Value'})
        df['topic'] = [topic]*len(df)
        topic_trends = topic_trends.append(df)

topic_trends.to_csv('stable_topics.csv')
```

### Read the data

```python
# read data, parsing the column "date" as a date
df = pd.read_csv('stable_topics.csv', 
                 parse_dates=['date'])
                 # make sure that records are sorted by date
```

## Checking time series stationarity

### Distributions of time series data

In the following we visualize first the distribution of trends data. If we observe that it is normally distributed and not multimodal we have a first indicative that the time series mean is constant with a finite variance.

```python
for topic in df.topic.unique():
    df_filt = df[(df['topic']==topic)]
    fig, axes = joypy.joyplot(df_filt, by="Region", column="Value", ylabels=True, xlabels=True, range_style='own', 
                            grid=False, linewidth=1, legend=False, figsize=(5,6),
                            title=f"Region {topic} trends",
                           )
```

### Plotting rolling means and variances 

Next we plot the rolling mean and variance for each timeseries for each topic and region. We smooth using a yearly window size (this could also be substituted for a 1 month window.) 
If the rolling mean and variance are approximately constant then the timeseries is stationary.

```python
for topic in df.topic.unique():
    for reg in df.Region.unique():
        df_filt = df[(df['Region']==reg) & (df['topic']==topic)]
        df_plot = df_filt[['date','Value']].dropna()
        df_avg = df_filt['Value'].dropna().rolling(window=53,min_periods=1,center=True).mean()
        df_std = df_filt['Value'].dropna().rolling(window=53,min_periods=1,center=True).var()

        # plotting transformed time series and rolling avg and std
        fig,ax = plt.subplots(1,1,figsize=(10,3))
        ax.plot(df_plot['date'],df_plot['Value'],'-',linewidth=1,label='popularity',color='#FFC3C3')
        ax.plot(df_plot['date'],df_avg,'--',label='rolling average',color='#FF985A')
        ax.plot(df_plot['date'],df_std,'-.',label='rolling variance',color='#7CB86A')
        #ax.set_ylabel('Outdoors & Recreation visits')
        ax.set_ylim(0,500)
        ax.set_xlabel('Time')
        plt.legend()
        plt.tight_layout()
        plt.title(f'{reg}_{topic}')
        plt.savefig(f'figs/{reg}_{topic}.png', dpi=400)
```

### Statistical tests of unit roots

We run the Adfuller-test to test for the existence of a unit root and the kpss-test to test for trend-stationarity. If both tests seem to indicate stationarity the timeseries is most likely stationary.

```python

# statistical tests to check for stationarity
for topic in df.topic.unique():
    for reg in df.Region.unique():
        df_filt = df[(df['Region']==reg) & (df['topic']==topic)]
        df_plot = df_filt[['date','Value']].dropna()
        print(f'For {topic} in Region {reg}:')
        print(" > Is the data stationary ?")
        dftest = adfuller(df_plot['Value'], autolag='AIC')
        print("Test statistic = {:.8f}".format(dftest[0]))
        print("P-value = {:.8f}".format(dftest[1]))
        print("Critical values :")
        for k, v in dftest[4].items():
            print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
            
        print(" > Is the data stationary ?")
        kpss_test = kpss(df_plot['Value'], 'c', 53)
        print("Test statistic = {:.8f}".format(kpss_test[0]))
        print("P-value = {:.8f}".format(kpss_test[1]))
        for key, value in kpss_test[3].items():
            print(f'   {key} : {value}')
        #print(f'Result: The series is {"not " if kpss_test[1] < 0.05 else ""}stationary')
        
        print('\n')
```

The test results indicate that the sports trends in particular are trend-stationary.

### Autocorrelation and partial autocorrelation plots

```python
%%capture --no-display
plt.style.use('ggplot')
# autocorrelation plots
for topic in df.topic.unique():
    for reg in df.Region.unique():
        df_filt = df[(df['Region']==reg) & (df['topic']==topic)]
        df_plot = df_filt[['date','Value']].dropna()
        fig, axs = plt.subplots(1, 2, figsize=(10,6))
        fig.suptitle(f'{reg}_{topic}')
        plot_acf(df_plot['Value'], ax=axs[0], lags=54)
        plot_pacf(df_plot['Value'], ax=axs[1], lags=54)
        axs[1].set_xlabel('Lag')
        axs[0].set_xlabel('Lag')
        plt.tight_layout()
        plt.savefig(f'figs/autocorrelation_{reg}_{topic}.png', dpi=400)
```

The autocorrelation and partial autocorrelation plots indicate that most of the time series only show significant autocorrelation for the first 3-4 lags except movies which has yearly autocorrelation. 

However the plots along with the unit root testing suggests that sports would be an reasonably stable time series to use as reference
